```python
import numpy as np
from matplotlib import pyplot as plt
import os
import pandas as pd
from collections import OrderedDict
```

```python
train_path = '../input/data/train'
train_file = 'train.csv'

df = pd.read_csv(os.path.join(train_path, train_file))
gender = df.value_counts('gender')
gender_dict = dict()
for index, value in gender.items():
    gender_dict[index] = value

print(type(gender.keys()))

print(gender.index[0], gender[0])
print(gender.index[1], gender[1])

age = df.value_counts('age')
age_dict = dict()
for index, value in age.items():
    age_dict[index] = value
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

x = gender_dict.keys()
y = gender_dict.values()
print(x)
print(y)

ax.bar(x, y)

plt.show()
```

```python
fig2, axes2 = plt.subplots(1, 2, figsize=(12, 7))

labels = age_dict.keys()
values = age_dict.values()
# explode = [0, 0.1]

axes2[0].pie(values, labels=labels)
axes2[1].bar(labels, values)

plt.show()
```

```python
is_under_30 = df['age'] < 30
is_over_60 = df['age'] >= 60
is_between_30_and_60 = ~is_under_30 & ~is_over_60

is_male = df['gender'] == 'male'
is_female = df['gender'] == 'female'

male_under_30 = df[is_under_30 & is_male]
female_under_30 = df[is_under_30 & is_female]
male_between_30_and_60 = df[is_between_30_and_60 & is_male]
female_between_30_and_60 = df[is_between_30_and_60 & is_female]
male_over_60 = df[is_over_60 & is_male]
female_over_60 = df[is_over_60 & is_female]

feature_list = [male_under_30, female_under_30, male_between_30_and_60,
               female_between_30_and_60, male_over_60, female_over_60]
```

```python
fig, axes = plt.subplots(figsize=(12, 7))
fig.set_facecolor('white')

values = [len(feature) for feature in feature_list]
labels = ['male_under_30', 'female_under_30', 'male_between_30_and_60',
               'female_between_30_and_60', 'male_over_60', 'female_over_60']


axes.pie(values, labels=labels)
plt.show()
```
