```python
class Test_Df():
    def __init__(self,csv,df,col):
        self.csv = csv
        self.df = pd.DataFrame(columns=col)
    
    def fill(self):
        cnt = 0
        path = 'images/'
        sort_dir = sorted(os.listdir(path))[2700:]
        #print(sort_dir)
        for i in tqdm.tqdm(range(len(sort_dir))):
            
            directory = sort_dir[i]
            #print(directory)
            #print(directory)
            if directory[1] != '_':
                #print(directory)
                sort_fname = sorted(os.listdir(path + directory + '/'))[7:]
                #print(sort_dic)
                
                for f in sort_fname:
                    if f[0] != '.':
                        #print(f)
                        self.df.loc[cnt] = csv_.loc[i][['gender','age']]
                        self.df.loc[cnt]['fname'] = f
                        self.df.loc[cnt]['path'] = directory
                        self.fill_label(self.df,i,cnt)
                        cnt += 1
        return self.df
    
    def fill_label(self,df,i,cnt):
        if 'incorrect' in df.loc[cnt]['fname']:
            if df.loc[cnt]['gender'] == 'male':
                if df.loc[cnt]['age'] < 30:
                    df.loc[cnt]['label'] = 6
                elif df.loc[cnt]['age'] < 60:
                    df.loc[cnt]['label'] = 7
                else:
                    df.loc[cnt]['label'] = 8
            else:
                if df.loc[cnt]['age'] < 30:
                    df.loc[cnt]['label'] = 9
                elif df.loc[cnt]['age'] < 60:
                    df.loc[cnt]['label'] = 10
                else:
                    df.loc[cnt]['label'] = 11
        elif 'mask' in df.loc[cnt]['fname']:
            if df.loc[cnt]['gender'] == 'male':
                if df.loc[cnt]['age'] < 30:
                    df.loc[cnt]['label'] = 0
                elif df.loc[cnt]['age'] < 60:
                    df.loc[cnt]['label'] = 1
                else:
                    df.loc[cnt]['label'] = 2
            else:
                if df.loc[cnt]['age'] < 30:
                    df.loc[cnt]['label'] = 3
                elif df.loc[cnt]['age'] < 60:
                    df.loc[cnt]['label'] = 4
                else:
                    df.loc[cnt]['label'] = 5
        elif 'normal' in df.loc[cnt]['fname']:
            if df.loc[cnt]['gender'] == 'male':
                if df.loc[cnt]['age'] < 30:
                    df.loc[cnt]['label'] = 12
                elif df.loc[cnt]['age'] < 60:
                    df.loc[cnt]['label'] = 13
                else:
                    df.loc[cnt]['label'] = 14
            else:
                if df.loc[cnt]['age'] < 30:
                    df.loc[cnt]['label'] = 15
                elif df.loc[cnt]['age'] < 60:
                    df.loc[cnt]['label'] = 16
                else:
                    df.loc[cnt]['label'] = 17
        
                    
                
            
        
    
    
```
